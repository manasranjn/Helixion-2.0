```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.plot(x, y)
```

```python
plt.plot(x, y, color='r')
```

```python
plt.plot(x, y, color='r', linestyle='--')
```

```python
plt.plot(x,y, 'r+')
```

```python
plt.plot(x,y, 'r+--')
```

```python
plt.plot(x, y, color='r', linestyle='-.')
```

```python
plt.plot(x, y, color='r', linestyle=':')
```

```python
plt.plot(x, y, color='r', linestyle='--', marker='o')
```

```python
plt.plot(x, y, color='r', linestyle='--', marker='s')
```

```python
plt.plot(x, y, color='r', linestyle='--', marker='x')
```

```python
plt.plot(x, y, color='g', linestyle='--', marker='x', markersize=15)
```

```python
plt.plot(x, y, color='b', linestyle='--', marker='x' , linewidth=4)
```

```python
plt.plot(x, y, color='b', linestyle='--', marker='x' , linewidth=4, alpha=1)
```

```python
plt.plot(x, y, color='b', linestyle='--', marker='x' , linewidth=4, alpha=0.4)
```

```python
plt.plot(x, y, color='b', linestyle='--', marker='x' , linewidth=4)
plt.title("Simple Line Plot")
```

```python
plt.plot(x, y, color='b', linestyle='--', marker='x' , linewidth=4)
plt.title("Simple Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
```

```python
plt.plot(x, y, color='b', linestyle='--', marker='D' , markersize=20)
plt.title("Simple Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
```

```python
plt.plot( [10, 20, 25, 30], label='Sales 2024')
plt.plot( [15, 18, 22, 28], label='Sales 2025')
plt.legend()
```

```python
plt.plot( [10, 20, 25, 30], label='Sales 2024')
plt.plot( [15, 18, 22, 28], label='Sales 2025')
plt.legend(loc='best', shadow=True)
```

```python
plt.plot([1,2,3],[3,4,5], label='Line A')
plt.plot([1,2,3],[5,3,1], label='Line B')

plt.legend(loc='upper left', title='Legend Title', fontsize=10, frameon=True)
```

# Bars/Pi chart

```python
x = ['A', 'B', 'C', 'D']
y = [10, 15, 7, 12]

plt.bar(x, y, color='skyblue')
plt.title("Bar Chart")
```

```python
company = ['TCS', 'RPW', 'OLA', 'JPW']
revenue = [123000, 156000, 90000, 105000]
comp = np.arange(len(company))
plt.bar(comp,revenue)
```

```python
plt.barh(x, y, color='orange')
```

```python
x = [1, 2, 3, 4, 5]
y = [5, 7, 4, 6, 8]

plt.scatter(x, y, color='green')
plt.title("Scatter Plot")
```

```python
data = np.random.randn(1000)
plt.hist(data, bins=20, color='purple', edgecolor='black')
plt.title("Histogram")
```

```python
labels = ['Python', 'Java', 'C++', 'Ruby']
sizes = [40, 30, 20, 10]
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title("Programming Language Usage")
```
