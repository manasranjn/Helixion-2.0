# Bars/Pi chart

```python
x = ['A', 'B', 'C', 'D']
y = [10, 15, 7, 12]

plt.bar(x, y, color='skyblue')
plt.title("Bar Chart")
```

```python
company = ['TCS', 'RPW', 'OLA', 'JPW']
revenue = [123000, 156000, 90000, 105000]
comp = np.arange(len(company))
plt.bar(comp,revenue)
```

```python
plt.xticks(comp,company)
plt.bar(comp,revenue)
```

```python
plt.xticks(comp,company)
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("Companies Revenue")
plt.bar(comp,revenue, label='Revenue')
plt.legend()
```

```python
plt.xticks(comp,company)
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("Companies Revenue")
plt.bar(comp,revenue, label='Revenue')

profit=[93000, 86000, 24000, 55000]
plt.bar(comp,profit,label='Profit')
plt.legend()
```

```python
plt.xticks(comp,company)
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("Companies Revenue")
plt.bar(comp-0.2,revenue, width=0.3, label='Revenue')

profit=[93000, 86000, 24000, 55000]
plt.bar(comp+0.1,profit, width=0.3, label='Profit')
plt.legend()
```

```python
labels = ['Python', 'Java', 'C++', 'Ruby']
sizes = [40, 30, 20, 10]

plt.pie(sizes, labels=labels)
plt.title("Programming Language Usage")
```
