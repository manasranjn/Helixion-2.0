```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.plot(x, y)
```

```python
plt.plot(x, y, label="Line 1")
```

```python
plt.plot(x, y, color='r')
```

```python
plt.plot(x, y, color='r', linestyle='--')
```

```python
plt.plot(x, y, color='r', linestyle='-.')
```

```python
plt.plot(x, y, color='r', linestyle=':')
```

```python
plt.plot(x, y, color='r', linestyle='--', marker='o')
```

```python
plt.plot(x, y, color='r', linestyle='--', marker='s')
```

```python
plt.plot(x, y, color='r', linestyle='--', marker='x')
```

```python
plt.plot(x, y, color='b', linestyle='--', marker='x' , linewidth=4)
```

```python
plt.plot(x, y, color='b', linestyle='--', marker='x' , linewidth=4)
plt.title("Simple Line Plot")
```

```python
plt.plot(x, y, color='b', linestyle='--', marker='x' , linewidth=4)
plt.title("Simple Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
```
