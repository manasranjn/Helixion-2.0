```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.plot(x, y)
```

```python
plt.plot(x, y, color='g')
```

```python
plt.plot(x, y, color='r', linestyle='--')
```

```python
plt.plot(x, y, color='g', linestyle='--', marker='s')
```

```python
plt.plot(x,y, ':r+')
```

```python
plt.plot(x,y, 'bx--')
```

```python
plt.plot(x, y, color='g', linestyle='--', marker='x', markersize=25)
```

```python
plt.plot(x, y, color='b', linestyle='-.', marker='o', markersize=10 , linewidth=4)
```

```python
plt.plot(x, y, color='b', linestyle='--', marker='x' , linewidth=4, alpha=1)
```

```python
plt.plot(x, y, color='b', linestyle='--', marker='x' , linewidth=4, alpha=0.4)
```

```python
plt.plot(x, y, color='b', linestyle='--', marker='x' , linewidth=4)
plt.title("Simple Line Plot")
```
