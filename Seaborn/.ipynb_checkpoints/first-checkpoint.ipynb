```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load an example dataset
df = sns.load_dataset("tips")
df.head()
```

```python
sns.barplot(y = 'total_bill', x ='size', data=df)
```

```python
sns.barplot(y = 'total_bill', x ='size', data=df, color='green')
```

```python
sns.barplot(y = 'total_bill', x ='size', data=df, palette='Set2')
```

```python
sns.barplot(x = 'size', y ='tip', data=df, palette='rocket')
```

```python
sns.countplot(x = 'size', data=df, palette='Set1')
```

```python
sns.countplot(y = 'size', data=df, palette='Set1')
```

```python
sns.countplot(x = 'tip', hue='size', data=df, palette='Set1')
```

```python
sns.displot(df.tip,  bins=5, color='b')
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.histplot(df["total_bill"], kde=True, bins=20)
plt.title("Distribution of Total Bill")
```

```python
sns.kdeplot(df["tip"], shade=True)
plt.title("Density Plot of Tips")
```

```python
sns.displot(df["total_bill"], kind="kde")
```

```python
sns.scatterplot(x="total_bill", y="tip", data=df, hue="sex", style="smoker")
plt.title("Total Bill vs Tip")
```

```python
sns.lineplot(x="size", y="tip", data=df)
plt.title("Line Plot between Size and Tip")
```

```python
sns.relplot(x="total_bill", y="tip", hue="day", kind="scatter", data=df)
```

```python
sns.boxplot(x="day", y="total_bill", data=df, hue="smoker")
plt.title("Box Plot of Total Bill by Day and Smoking Status")
```
