```python
def convert_gender_cell(cell):
    if cell=="Male":
        return 'Man'
    return cell

def convert_Country_cell(cell):
    if cell=="United States":
        return "US"
    return cell

df = pd.read_excel("file_example.xls", "Sheet2", converters = {
'Gender': convert_gender_cell,
'Country': convert_Country_cell})

df
```

```python
df.to_excel("newData.xlsx", sheet_name="data")
```
