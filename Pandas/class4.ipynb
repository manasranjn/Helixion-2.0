# Handling Missing Data

```python
import pandas as pd
df = pd.read_csv("customers-100.csv")
df
```

```python
type(df['Subscription Date'][0])
```

```python
df = pd.read_csv('customers-100.csv', parse_dates=['Subscription Date'])
type(df['Subscription Date'][0])
```

```python
df2 = df.fillna("0")
df2.head(10)
```

```python
df3 = df.fillna({
    'Last Name': "No Last name",
    'Company': "No Company",
    'City': "No City",
    'Country': "No Country"
})
df3
```

```python
df.head()
```

```python
df4 = df.fillna(method='ffill')
df4
```

```python
df4 = df.fillna(method='bfill')
df4.head(10)
```

```python
df6 = df.fillna(method='ffill', axis='columns')
df6.head(7)
```

```python
df6 = df.fillna(method='bfill', axis='columns')
df6.head()
```

```python
df7 = df.dropna()
df7.head()
```

```python
df8 = df.dropna(how="all")
df8.head()
```
