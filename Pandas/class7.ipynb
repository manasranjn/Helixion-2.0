```python
import pandas as pd
weather1 = pd.DataFrame({
    "City": ['Mumbai', 'Bhubaneswar','Bengaluru'],
    "temp": [12,13,19],
    "wind": [5,6,23]})
weather1
```

```python
weather2 = pd.DataFrame({
    "City": ['Chennai', 'Delhi','Hyderabad'],
    "temp": [23,23,29],
    "wind": [5,6,9]})
weather2
```

```python
df = pd.concat([weather1, weather2], ignore_index=True)
df
```

```python
df = pd.concat([weather1, weather2], keys=["DF1", "DF2"])
df
```

```python
df.loc['DF2']
```

```python
df = pd.concat([weather1, weather2], axis= 1)
df
```

```python
df1 = pd.DataFrame({
    "City": ['Mumbai', 'Bhubaneswar','Bengaluru'],
    "temp": [12,13,19],
    "wind": [5,6,23]}, index = [1,2,3])
df1
```

```python
temp = pd.DataFrame({
    "City": ['Mumbai', 'Bhubaneswar','Bengaluru'],
    "temp": [12,13,19]})
temp
```

```python
wind = pd.DataFrame({
    "City": ['Mumbai', 'Bhubaneswar','Delhi'],
    "wind": [5,6,23]})
wind
```

```python
df = pd.merge(temp, wind, on='City')
df
```

```python
df = pd.merge(temp, wind, on='City', how="left")
df
```

```python
df = pd.merge(temp, wind, on='City', how="right")
df
```

```python
df = pd.merge(temp, wind, on='City', how="outer")
df
```

```python
df = pd.merge(temp, wind, on='City', how="outer", indicator=True)
df
```
