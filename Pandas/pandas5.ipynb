```python
import pandas as pd

# Create sample DataFrame
data = {
    'Employee': ['A', 'B', 'C', 'A', 'B', 'C'],
    'Department': ['HR', 'HR', 'HR', 'IT', 'IT', 'IT'],
    'Year': [2022, 2022, 2022, 2023, 2023, 2023],
    'Salary': [40000, 42000, 41000, 45000, 47000, 46000],
    'Bonus': [2000, 2100, 1900, 2500, 2700, 2600]
}

df = pd.DataFrame(data)
df
```

```python
pivot_df = df.pivot(index='Employee', columns='Year', values='Salary')
pivot_df
```

```python
pivot_table_df = df.pivot_table(
    index='Employee',
    columns='Year',
    values='Salary',
)
pivot_table_df
```

```python
pivot_table_df = df.pivot_table(
    index='Employee',
    columns='Year',
    values='Salary',
    aggfunc='sum',     
    fill_value=0,      
    margins=True,    
    margins_name='Total'
)
pivot_table_df
```

```python
melt_df = pd.melt(
    df,
    id_vars=['Employee', 'Department'],   
    value_vars=['Salary', 'Bonus'],       
    var_name='New Col',                  
    value_name='Amount'                   
)
melt_df
```

```python
df.rename(columns={'Employee':'Emp Name'}, inplace=True)
df
```
